```python
# Text Pre-Processing
```

```python
# Using Regular Expression
import re
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords

def preprocess(sentence):
    sentence = str(sentence)                    # Convert input to string
    sentence = sentence.lower()                 # Lowercase all text (normalize words)
    sentence = sentence.replace('{html}',"")    # Remove the literal "{html}" tag
    
    cleanr = re.compile('<.*?>')                # Regex pattern to find HTML tags
    cleantext = re.sub(cleanr, '', sentence)    # Remove HTML tags like <p>, <div>
    
    rem_url = re.sub(r'http\S+', '', cleantext) # Remove URLs (http://, https://, etc.)
    rem_num = re.sub('[0-9]+', '', rem_url)     # Remove numbers (123, 2025, etc.)
    
    tokenizer = RegexpTokenizer(r'\w+')         # Tokenizer: keeps only words (no punctuation)
    tokens = tokenizer.tokenize(rem_num)        # Split sentence into word tokens
    
    # Remove stopwords (like "the", "is", "and") and keep words with length > 2
    filtered_words = [w for w in tokens if len(w) > 2 if not w in stopwords.words('english')]
    
    return " ".join(filtered_words)             # Join tokens back into a cleaned string
```

```python
# Model Building
```

```python
df = pd.read_csv("Cleaned_data.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['category'] = le.fit_transform(df['Category'])
```

```python
df.head()
```

```python
df.describe().T
```

```python
df.isnull().sum()
```

```python
x = df['text'].values
y = df['category'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, stratify=y)
x_train.shape, x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(sublinear_tf=True,stop_words='english')
x_train = tfv.fit_transform(x_train)
x_test = tfv.transform(x_test)
```

```python
x_train.shape, x_test.shape
```

```python
print(y_train), print(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "SVM": SVC(kernel="linear"),
    "Naive Bayes": MultinomialNB(),
    "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42)
}
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

results = {}

for name, model in models.items():
    print(f'\nTraining {name}...')

    model.fit(x_train, y_train)

    y_pred = model.predict(x_test)

    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"Test Accuracy ({name}): {acc: .4f}")

    print(f"Classification Report - {name}:\n")
    print(classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
    plt.figure(figsize=(10,6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=model.classes_, yticklabels=model.classes_)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title(f"Confusion Matrix - {name}")
    plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=list(results.keys()), y=list(results.values()), palette="plasma")
plt.ylabel("Accuracy")
plt.title("Model Comparison - Resume Classification")
plt.xticks(rotation=30)
plt.show()
```
